```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import spacy
```

```python
df = pd.DataFrame()
for i in range(1, 10):
    df = df.append(pd.read_csv("IRAhandle_tweets_" + str(i) + ".csv"))
    print("reading csv " + str(i) + "/9")
```

```python
categories = df.groupby('account_category')
```

```python
sizes = categories.size()
```

```python
sums = categories.sum()
```

```python
sums.insert(len(sums.columns), 'sizes', sizes)
```

```python
sums.insert(len(sums.columns), 'interaction ratio', sums['updates']/sums['sizes'])
```
